```python
import pandas as pd
import matplotlib.pyplot as mtb
import seaborn as sb

data = pd.read_csv("sales.csv")

# data.head(10)
```

```python
sales = data.query("Sales>1000")["Sales"].to_frame()
```

```python
mtb.figure(figsize=(18, 6))
sb.histplot(sales, bins=100, color="red")
mtb.title("Distribution of Sales")
mtb.xlabel("Sales")
mtb.ylabel("Distribution")
mtb.grid(axis="y", linestyle="--", alpha=.7)
mtb.show()
```

```python
top_sc = data.groupby("Sub-Category")["Profit"].sum().sort_values(ascending=False).head(10)
```

```python
mtb.figure(figsize=(12, 6))
sb.barplot(x=top_sc.index, y=top_sc.values, color="red")
mtb.title("Top 10 Sub-Categories by Profit")
mtb.xlabel("Sub-Category")
mtb.ylabel("Profit")
mtb.grid(axis="y", linestyle="--", alpha=.7)
mtb.show()
```

```python
sales_by_sp = data.groupby("Country/Region")["Sales"].sum()
```

```python
mtb.figure(figsize=(12, 6))
sb.barplot(x=sales_by_sp.index, y=sales_by_sp.values, color="blue")
mtb.title("Sales by Country/Region")
mtb.xlabel("Country/Region")
mtb.ylabel("Sales")
mtb.grid(axis="y", linestyle="--", alpha=.7)
mtb.show()
```
