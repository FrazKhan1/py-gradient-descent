```python
import pandas as pd
import matplotlib.pyplot as mtb
import seaborn as sb

data = pd.read_csv("sales.csv")

# data.head(10)
```

```python
sales = data["Sales"]
```

```python
sales
```

```python
mtb.figure(figsize=(24, 6))
sb.histplot(sales, bins=100, color="blue")
mtb.title("Distribution of Sales")
mtb.xlabel("Sales")
mtb.ylabel("Distribution")
mtb.grid(axis="y", linestyle="--", alpha=0.5)
mtb.show()
```
