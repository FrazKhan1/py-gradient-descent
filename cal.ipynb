```python
import numpy as np

n=10
w= np.random.randint(1,100,size=n)
print(w)
x = np.random.randint(1,100, size=n)
print(x)
h_score = np.random.randint(1,100,size=n)
print(h_score)



loses =[]

for i in range(len(w)):
    for j in range(len(x)):
        y = (w[i]*x[j]) - h_score[j]
        print(y)
        loses.append(y)
   
cost = np.sum(loses) / len(loses)
print(cost) 
```

```python
data = np.random.rand(20, 2)
X = data[:, 0]
Y = data[:, 1]
print("data",data)
print(X)

weights = np.array([0.5, 0.54, 0.004,0.786])
b = 2

for m in weights:
    Y_pred = m * X + b
    loss = 0.5*(Y - Y_pred)**2
    cost = np.mean(loss)
    print(f"Weight: {m}, Loss per row: {loss}, Total Cost: {cost}")
```
