```python
np_array = np.random.rand(10, 10)
np_array
```

```python
first = np_array[0]
first
```

```python
nine_rows = np_array[1:]
nine_rows
```

```python
nine_rows[0]
```

```python
  
results = []
  
for i in range(len(nine_rows)):
    res = nine_rows[i]
    dot_p = np.dot(first,res)
    mod = norm(first) + norm(res)
    cosine = dot_p / mod
    results.append(cosine * 100)
    
results
```
